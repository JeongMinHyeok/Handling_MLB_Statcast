## 블게주와 오타니의 주요 지표 비교

### 지표 설명
- xBA : 기대타율
- xSLG : 기대장타율
- xwOBA : 기대 가중출루율(평균) (공격기여도)
- xOBP : 기대출루율
- xISO : 기대순수장타율
- exit_velocity_avg (EVA) : 평균 타구속도
- launch_angle_avg (LAA) : 평균 타구각도
- sweet_spot_percent (SSP) : 타구각도가 8도에서 32도 사이인 타구의 비율
- barrel_batted_rate (BBR) : 배럴타구(sweet spot & over 95mph) 비율
- hard_hit_percent (HHP) : 강한타구(95마일 이상)의 비율
